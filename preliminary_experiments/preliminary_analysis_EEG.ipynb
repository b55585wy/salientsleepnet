```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# 数据目录
# 常见的频段包括：

# Delta (0.5-4 Hz)
# Theta (4-8 Hz)
# Alpha (8-13 Hz)
# Beta (13-30 Hz)
directory = 'F:/sleep-edf-database-expanded-1.0.0/sleep-edf-database-expanded-1.0.0/filtered_single_channel_data'

# 读取原始数据
W = np.load(os.path.join(directory , 'filtered_W.npy'))
N1 = np.load(os.path.join(directory , 'filtered_N1.npy'))
N2 = np.load(os.path.join(directory , 'filtered_N2.npy'))
N3 = np.load(os.path.join(directory , 'filtered_N3.npy'))
REM = np.load(os.path.join(directory , 'filtered_REM.npy'))

# 查看数据的基本信息
print(f'W shape: {W.shape}')
print(f'N1 shape: {N1.shape}')
print(f'N2 shape: {N2.shape}')
print(f'N3 shape: {N3.shape}')
print(f'REM shape: {REM.shape}')

# 计算各个阶段的平均值和标准差
W_mean = np.mean(W)
W_std = np.std(W)
N1_mean = np.mean(N1)
N1_std = np.std(N1)
N2_mean = np.mean(N2)
N2_std = np.std(N2)
N3_mean = np.mean(N3)
N3_std = np.std(N3)
REM_mean = np.mean(REM)
REM_std = np.std(REM)

# 打印统计信息
print(f'W stage - Mean: {W_mean}, Std: {W_std}')
print(f'N1 stage - Mean: {N1_mean}, Std: {N1_std}')
print(f'N2 stage - Mean: {N2_mean}, Std: {N2_std}')
print(f'N3 stage - Mean: {N3_mean}, Std: {N3_std}')
print(f'REM stage - Mean: {REM_mean}, Std: {REM_std}')

# 定义计算功率谱密度和频段功率的函数
def compute_psd_and_band_power(data, stage, sfreq=100, nperseg=1000):
    # 计算功率谱密度
    f, Pxx = welch(data.flatten(), sfreq, nperseg=nperseg)
    
    # 绘制功率谱密度
    plt.figure(figsize=(10, 6))
    plt.semilogy(f, Pxx)
    plt.title(f'Power Spectral Density - {stage}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power/Frequency (dB/Hz)')
    plt.grid()
    plt.show()
    
    # 计算各频段的功率
    delta_power = np.trapz(Pxx[(f >= 0.5) & (f < 4)], f[(f >= 0.5) & (f < 4)])
    theta_power = np.trapz(Pxx[(f >= 4) & (f < 8)], f[(f >= 4) & (f < 8)])
    alpha_power = np.trapz(Pxx[(f >= 8) & (f < 13)], f[(f >= 8) & (f < 13)])
    beta_power = np.trapz(Pxx[(f >= 13) & (f < 30)], f[(f >= 13) & (f < 30)])
    
    return delta_power, theta_power, alpha_power, beta_power

# 计算和可视化每个阶段的功率谱密度和频段功率
for stage, data in [('W', W), ('N1', N1), ('N2', N2), ('N3', N3), ('REM', REM)]:
    delta_power, theta_power, alpha_power, beta_power = compute_psd_and_band_power(data, stage)
    print(f'\n{stage} stage band powers:')
    print(f'Delta Power: {delta_power}')
    print(f'Theta Power: {theta_power}')
    print(f'Alpha Power: {alpha_power}')
    print(f'Beta Power: {beta_power}')
```

```python
import os
import numpy as np
import pywt
import matplotlib.pyplot as plt

# 定义小波分解和重构函数
def wavelet_decompose_reconstruct(data, wavelet='db4', level=4):
    # 执行小波分解
    coeffs = pywt.wavedec(data, wavelet, level=level)
    
    # 重构不同频段的信号
    cA4 = pywt.waverec([coeffs[0]] + [None] * level, wavelet)
    cD4 = pywt.waverec([None] * 1 + [coeffs[1]] + [None] * (level - 1), wavelet)
    cD3 = pywt.waverec([None] * 2 + [coeffs[2]] + [None] * (level - 2), wavelet)
    cD2 = pywt.waverec([None] * 3 + [coeffs[3]] + [None] * (level - 3), wavelet)
    
    return cA4, cD4, cD3, cD2

# 加载数据（示例以 filtered_W.npy 为例）
directory = 'F:/sleep-edf-database-expanded-1.0.0/sleep-edf-database-expanded-1.0.0/filtered_single_channel_data'
W = np.load(os.path.join(directory , 'filtered_W.npy'))

# 选择一个样本进行小波分解和重构
sample_data = W[0, 0, :]  # 选择第一个样本的第一个通道

# 进行小波分解和重构
delta_wave, theta_wave, alpha_wave, beta_wave = wavelet_decompose_reconstruct(sample_data)

# 绘制原始信号和不同频段的分离信号
plt.figure(figsize=(12, 10))

plt.subplot(5, 1, 1)
plt.plot(sample_data)
plt.title('Original Signal')

plt.subplot(5, 1, 2)
plt.plot(delta_wave)
plt.title('Delta Wave (0-4 Hz)')

plt.subplot(5, 1, 3)
plt.plot(theta_wave)
plt.title('Theta Wave (4-8 Hz)')

plt.subplot(5, 1, 4)
plt.plot(alpha_wave)
plt.title('Alpha Wave (8-13 Hz)')

plt.subplot(5, 1, 5)
plt.plot(beta_wave)
plt.title('Beta Wave (13-40 Hz)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

def max_distance(x_i, x_j):
    return np.max(np.abs(x_i - x_j))

def phi(m, r, data):
    N = len(data)
    X = np.array([data[i:i + m] for i in range(N - m + 1)])
    C = np.zeros(N - m + 1)
    
    for i in range(N - m + 1):
        for j in range(N - m + 1):
            if max_distance(X[i], X[j]) < r:
                C[i] += 1
        C[i] /= (N - m + 1)
        
    return np.sum(np.log(C)) / (N - m + 1)

def approximate_entropy(data, m, r):
    return abs(phi(m, r, data) - phi(m + 1, r, data))
import os
import numpy as np

# 数据目录
directory = 'F:/sleep-edf-database-expanded-1.0.0/sleep-edf-database-expanded-1.0.0/filtered_single_channel_data'

# 读取原始数据
W = np.load(os.path.join(directory , 'filtered_W.npy'))
N1 = np.load(os.path.join(directory , 'filtered_N1.npy'))
N2 = np.load(os.path.join(directory , 'filtered_N2.npy'))
N3 = np.load(os.path.join(directory , 'filtered_N3.npy'))
REM = np.load(os.path.join(directory , 'filtered_REM.npy'))

# 选择参数
m = 2
r = 0.2 * np.std(W)  # 可以根据数据的标准差调整阈值

# 计算近似熵
W_entropy = approximate_entropy(W[0, 0, :], m, r)
N1_entropy = approximate_entropy(N1[0, 0, :], m, r)
N2_entropy = approximate_entropy(N2[0, 0, :], m, r)
N3_entropy = approximate_entropy(N3[0, 0, :], m, r)
REM_entropy = approximate_entropy(REM[0, 0, :], m, r)

# 打印近似熵结果
print(f'W stage approximate entropy: {W_entropy}')
print(f'N1 stage approximate entropy: {N1_entropy}')
print(f'N2 stage approximate entropy: {N2_entropy}')
print(f'N3 stage approximate entropy: {N3_entropy}')
print(f'REM stage approximate entropy: {REM_entropy}')
```

```python
import numpy as np

def sample_entropy(data, m, r):
    N = len(data)
    def _max_dist(x_i, x_j):
        return np.max(np.abs(x_i - x_j))
    
    def _phi(m):
        X = np.array([data[i:i + m] for i in range(N - m + 1)])
        C = np.sum([np.sum([1 if _max_dist(X[i], X[j]) < r else 0 for j in range(N - m + 1)]) - 1 for i in range(N - m + 1)])
        return C / (N - m + 1) / (N - m)
    
    return -np.log(_phi(m + 1) / _phi(m))

import os
import numpy as np

# 数据目录
directory = 'F:/sleep-edf-database-expanded-1.0.0/sleep-edf-database-expanded-1.0.0/filtered_single_channel_data'

# 读取原始数据
W = np.load(os.path.join(directory , 'filtered_W.npy'))
N1 = np.load(os.path.join(directory , 'filtered_N1.npy'))
N2 = np.load(os.path.join(directory , 'filtered_N2.npy'))
N3 = np.load(os.path.join(directory , 'filtered_N3.npy'))
REM = np.load(os.path.join(directory , 'filtered_REM.npy'))

# 选择参数
m = 2
r = 0.2 * np.std(W)  # 可以根据数据的标准差调整阈值

# 计算样本熵
W_sampen = sample_entropy(W[0, 0, :], m, r)
N1_sampen = sample_entropy(N1[0, 0, :], m, r)
N2_sampen = sample_entropy(N2[0, 0, :], m, r)
N3_sampen = sample_entropy(N3[0, 0, :], m, r)
REM_sampen = sample_entropy(REM[0, 0, :], m, r)

# 打印样本熵结果
print(f'W stage sample entropy: {W_sampen}')
print(f'N1 stage sample entropy: {N1_sampen}')
print(f'N2 stage sample entropy: {N2_sampen}')
print(f'N3 stage sample entropy: {N3_sampen}')
print(f'REM stage sample entropy: {REM_sampen}')
```
